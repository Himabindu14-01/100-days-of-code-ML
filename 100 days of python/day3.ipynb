```python
x = 12
y = 4

print("Add:", x + y)
print("Sub:", x - y)
print("Mul:", x * y)
print("Div:", x / y)
print("Floor Div:", x // y)
print("Power:", x ** y)
print("Mod:", x % y)
```

```python
fruits = ["apple", "banana", "mango"]
print("List:", fruits)
print("First:", fruits[0])
print("Last:", fruits[-1])

fruits.append("orange")
print("After append:", fruits)

fruits.remove("banana")
print("After remove:", fruits)

print("Slice:", fruits[0:2])
```

```python
nums = [5, 2, 9, 4, 7]

total = 0
for n in nums:
    total += n

avg = total / len(nums)
print("Total:", total)
print("Avg:", avg)
```

```python
nums = [5, 2, 9, 4, 7]
rev = nums[::-1]
print("Reversed:", rev)
```
