```python
def stat(num):
    max_n = num[0]
    min_n = num[0]
    total = 0

    for n in num:
        if n < min_n:
            min_n = n
        if n > max_n:
            max_n = n
        
        total += n

    return min_n, max_n, total

num = [3, 5, 2, 7, 8]
print(stat(num))
```
